```
# 1. Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
from PyEMD import EMD
# Import the files module from google.colab to enable file uploads
from google.colab import files

# 2. Upload your audio file
# Use the files.upload() function to open a file picker
uploaded = files.upload()

# Get the filename of the uploaded file
# Assuming you upload only one file, get the first key from the dictionary
audio_filename = next(iter(uploaded))
print(f"Uploaded file: {audio_filename}")


# 3. Load the uploaded audio file
signal, sr = sf.read(audio_filename)   # Reads the audio file
if signal.ndim > 1:
    signal = signal[:, 0]  # Use only one channel if stereo

# 4. (Optional) Use only a segment for faster demo
# signal = signal[:sr*2]  # First 2 seconds

# 5. Apply EMD
emd = EMD()
imfs = emd.emd(signal)

print(f"Extracted {len(imfs)} IMFs")

# 6. Plot IMFs
plt.figure(figsize=(12, 2 * (len(imfs)+1)))
plt.subplot(len(imfs)+1, 1, 1)
plt.plot(signal, color='black')
plt.title('Original Signal')
for i, imf in enumerate(imfs):
    plt.subplot(len(imfs)+1, 1, i + 2)
    plt.plot(imf, color='b')
    plt.title(f'IMF {i + 1}')
plt.tight_layout()
plt.show()
```
